# Imports

```python
import pandas as pd

datafile = 'metoo_tweets_dec2017.xlsx'
#create data frames
df = pd.read_excel(datafile, usecols="B")
pd.set_option('display.max_colwidth', -1)

print("Number of tweets: ",len(df['text']))

df
```

# Data Cleaning and Data Pre-processing

```python
import re #regular expression

for i in range(len(df)):
    #replace each non-letter with empty string
    df.iloc[i, 0] = re.sub('@[^\s]+','',str(df.iloc[i, 0]))
    df.iloc[i, 0] = re.sub('#[^\s]+','',str(df.iloc[i, 0]))
    df.iloc[i, 0] = re.sub('http[^\s]+','',str(df.iloc[i, 0]))
    df.iloc[i, 0] = re.sub('https[^\s]+','',str(df.iloc[i, 0]))
    df.iloc[i, 0] = re.sub(r'[0-9]', '_', str(df.iloc[i, 0]))
```

```python
import pandas as pd
import re
import nltk
import nltk.corpus
from nltk.corpus import stopwords
from nltk.stem import WordNetLemmatizer
nltk.download('stopwords')
nltk.download('wordnet')

stop_words = stopwords.words('english') + ['say', 'someone' ,'get', 'go', 'know', 'may', 'need', 'like', 'make', 'see', 'want', 'come', 
                   'take', 'use', 'should', 'would', 'can','could', 'of', '#MeToo', 'many', '#metoo', 'one', 'two', 'who', 'rt', 'once', 
                   "don't", 'saying', 'every', 'image', 'at', 'The', 'your', 'I', 'Retweet', 'for', 'thing', 'notoritey', 
                   'time', 'how', 'de', 'en', 'caption', 'finally', 'also', 'copyright', 'something', 'to', 'guys', 'yet', 'way', 'jack', 'matt',
                   'moron', 'geraldo', 'garrisonamp', 'jr', 'thank', 'job', 'keillor', 'everyone', 'xdxd', 'puertoricorelief', 'harem', 
                   'megan', 'kerry', 'eye', 'jayz', 'xd', 'csa', 'lauer', 'mattlauer',  'youre', 'lineup', 'ask', 'nan', 'false'
                   'djt', 'jopis', 'xdmoralityxdhum', 'hoacoa', 'donburke', 'sundance', 'keira', 'knightley', 'daisy', 'ridley', 'matt lauer' 
                   'unsealthedeals', 'barbara', 'valter', 'nt', 'anita', 'hill', 'still', 'hes', 'hap', 'hell', 'im', 'actually', 'This', "It's", 'Im']
df['clean_text'] = df['text'].apply(lambda x: ' '.join([word for word in x.split() if word not in (stop_words)]))
df
```

```python
docs = list(df.loc[:, "clean_text"].values)
docs
```

# NMF

```python
from sklearn.feature_extraction.text import TfidfVectorizer
```

# Build the TF-IDF Matrix

```python
# Create an NMF instance: model
tfidf = TfidfVectorizer(max_df=0.95, min_df=2,stop_words='english')
```

```python
# Transform the TF-IDF: nmf_features
dm = tfidf.fit_transform(docs)
```

```python
dm
```

```python
from sklearn.decomposition import NMF
```

```python
nmf_model = NMF(n_components=7, random_state=42)
```

```python
# Fit the model to TF-IDF
nmf_model.fit(dm)
```

```python
tfidf.get_feature_names()
```

```python
#randome feature name
tfidf.get_feature_names()[200]
```

```python
for index,topic in enumerate(nmf_model.components_):
    print(f"THE TOP 15 WORDS FOR TOPIC # {index}")
    print([tfidf.get_feature_names()[i] for i in topic.argsort()[-30:]])
    print('\n')
```

```python
#tranform the result
topic_results = nmf_model.transform(dm)
```

```python
topic_results.argmax(axis=1)
```

```python
df['Topic'] = topic_results.argmax(axis=1)
```

```python
mytopic_dict = {0:'sexual',1:'harassment',2:'movement',3:'article',4:'behavior',5:'news',6:'accused'}
df['Topic Label'] = df['Topic'].map(mytopic_dict)
df
```
